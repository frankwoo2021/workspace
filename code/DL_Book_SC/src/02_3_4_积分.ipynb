# 积分(Integration)

## 范例1. 积分
### $\int x\;dx = \frac{1}{2} x^2$

```python
# 载入套件
import numpy as np
import scipy.integrate as integrate
import numpy as np
import math


f = lambda x: x 
i, e = integrate.quad(f, 0, 3)

print('积分值: ' + str(i))
print('误差: ' + str(e))
```

```python
import matplotlib.pyplot as plt

# 样本点
x = range(0, 11)

# 作图
plt.plot(x, f(x), color='purple')

# 积分面积
area = np.arange(0, 3, 1/20)
plt.fill_between(area, f(area), color='green')

# 设定图形属性
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()

plt.show()
```

## 范例2. 计算常态分配(-∞, ∞) 的机率。

```python
from IPython.display import Image
Image("./formula/Probability Density Function Of Normal Distribution.png")
```

```python
# 载入套件
import scipy.integrate as integrate
import numpy as np
import math


# 平均数(mean)、标准差(std)
mean = 0
std = 1
# 常态分配的机率密度函数(Probability Density Function, pdf)
f = lambda x: (1/((2*np.pi*std**2) ** .5)) * np.exp(-0.5*((x-mean)/std)**2) 

# 积分，从负无穷大至无穷大
i, e = integrate.quad(f, -np.inf, np.inf)

print('累积机率:', round(i, 2))
print('误差:', str(e))
```

## 计算常态分配的正负1、2、3倍标准差区间之机率

```python
# 1倍标准差区间之机率
i, e = integrate.quad(f, -1, 1)

print('累积机率:', round(i, 3))
print('误差:', str(e))
```

```python
# 2倍标准差区间之机率
i, e = integrate.quad(f, -2, 2)

print('累积机率:', round(i, 3))
print('误差:', str(e))
```

```python
# 3倍标准差区间之机率
i, e = integrate.quad(f, -3, 3)

print('累积机率:', round(i, 3))
print('误差:', str(e))
```

```python
# 95%信赖区间之机率
i, e = integrate.quad(f, -1.96, 1.96)

print('累积机率:', round(i, 3))
print('误差:', str(e))
```

## 使用随机乱数及直方图，绘制标准常态分配图

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# 使用 randn 产生标准常态分配的乱数 10000 个样本点
x = np.random.randn(10000)

# 直方图，参数 hist=False：不画阶梯直方图，只画平滑曲线
sns.distplot(x, hist=False)

# 设定图形属性
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()

# 对正负1、2、3倍标准差画虚线
plt.axvline(-3, c='r', linestyle=':')
plt.axvline(3, c='r', linestyle=':')
plt.axvline(-2, c='g', linestyle=':')
plt.axvline(2, c='g', linestyle=':')
plt.axvline(-1, c='b', linestyle=':')
plt.axvline(1, c='b', linestyle=':')

# 1 倍标准差机率
plt.annotate(text='', xy=(-1,0.25), xytext=(1,0.25), arrowprops=dict(arrowstyle='<->'))
plt.annotate(text='68.3%', xy=(0,0.26), xytext=(0,0.26))
# 2 倍标准差机率
plt.annotate(text='', xy=(-2,0.05), xytext=(2,0.05), arrowprops=dict(arrowstyle='<->'))
plt.annotate(text='95.4%', xy=(0,0.06), xytext=(0,0.06))
# 3 倍标准差机率
plt.annotate(text='', xy=(-3,0.01), xytext=(3,0.01), arrowprops=dict(arrowstyle='<->'))
plt.annotate(text='99.7%', xy=(0,0.02), xytext=(0,0.02))

plt.show()
```

## 使用SciPy计算标准常态分配之机率

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

z1, z2 = -4, 4
x = np.arange(z1, z2, 0.001) 
y = norm.pdf(x,0,1)

plt.plot(x,y,'black')
plt.fill_between(x,y,0, alpha=0.3, color='b')
```

## 计算常态分配的正负1、2、3倍标准差区间之机率

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

z1, z2 = -1, 1
x1 = np.arange(z1, z2, 0.001) 
y1 = norm.pdf(x1,0,1)

# 1倍标准差区域
plt.fill_between(x1,y1,0, alpha=0.3, color='b')
plt.ylim(0,0.5)

# 4倍标准差区域
z1, z2 = -4, 4
x2 = np.arange(z1, z2, 0.001) 
y2 = norm.pdf(x2,0,1)
plt.plot(x2,y2,'black')
plt.fill_between(x2,y2,0, alpha=0.1, color='b')
```
