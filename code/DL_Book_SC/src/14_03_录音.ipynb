## 载入相关套件

```python
# 载入相关套件
import speech_recognition as sr
import pyttsx3
```

## 列出电脑中的说话者(Speaker)

```python
# 列出电脑中的说话者(Speaker)
speak = pyttsx3.init()
voices = speak.getProperty('voices')
for voice in voices: 
    print("Voice:") 
    print(" - ID: %s" % voice.id) 
    print(" - Name: %s" % voice.name) 
    print(" - Languages: %s" % voice.languages) 
    print(" - Gender: %s" % voice.gender) 
    print(" - Age: %s" % voice.age)
```

```python
# 指定说话者
speak.setProperty('voice', voices[1].id)
```

```python
speak.say('No One Wants 40 Hour Work Weeks Anymore. Everyone Wants to Work 4 Hours Per Week on a Laptop In Bali')
# 等待说完
speak.runAndWait()
```

```python
# 指定说话者
speak.setProperty('voice', voices[0].id)
```

```python
speak.say('受峰面影响北台湾今天下午大雨特报，有些道路甚至发生积淹，曾文水库上游也传来好消息')
# 等待说完
speak.runAndWait()
```

```python
# 麦克风收音
# 受峰面影响北台湾今天下午大雨特报，有些道路甚至发生淹水
r = sr.Recognizer()
with sr.Microphone() as source:
    # 文字转语音
    speak.say('请说话...')
    # 等待说完
    speak.runAndWait()
    
    #降噪
    r.adjust_for_ambient_noise(source)
    # 麦克风收音
    audio = r.listen(source)
```

```python
# 录音存档    
wav_file = "./audio/woman.wav"
with open(wav_file, "wb") as f:
    f.write(audio.get_wav_data(convert_rate=16000))
```

```python
# 语音辨识
# 受峰面影响北台湾今天下午大雨特报有，些道路甚至发生积淹，曾文水库上游也传来好消息
try:
    text=r.recognize_google(audio, language='zh-tw')
    print(text)
except e:
    pass
```

## 取得音档的属性

```python
# 取得音档的属性
# https://docs.python.org/3/library/wave.html
import wave

f=wave.open(wav_file)
print(f'取样频率={f.getframerate()}, 帧数={f.getnframes()}, ' +
      f'声道={f.getnchannels()}, 精度={f.getsampwidth()}, ' +
      f'档案秒数={f.getnframes() / (f.getframerate() * f.getnchannels()):.2f}')
f.close()
```

```python
import speech_recognition as sr

# 读取音档，转为音讯
r = sr.Recognizer()
with sr.WavFile(wav_file) as source:
    audio = r.record(source)
    
# 语音辨识
try:
    text=r.recognize_google(audio, language='zh-tw')
    print(text)
except e:
    pass
```

```python
# 显示所有可能的辨识结果及信赖度
dict1=r.recognize_google(audio, show_all=True, language='zh-tw')
for i, item in enumerate(dict1['alternative']):
    if i == 0:
        print(f"信赖度={item['confidence']}, {item['transcript']}")
    else:
        print(f"{item['transcript']}")
```
