## 范例1. 绘制掷骰子的机率密度函数

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['Microsoft JhengHei'] 
plt.rcParams['axes.unicode_minus'] = False


# 点数与机率
face = [1,2,3,4,5,6]
probs = np.full((6), 1/6)

# 绘制长条图
plt.bar(face, probs)
plt.ylabel('机率', fontsize=12)
plt.xlabel('点数', fontsize=12)
plt.title('掷骰子', fontsize=12)
plt.ylim([0,1])
plt.show()
```

## 范例2. 绘制掷硬币的机率密度函数

```python
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['Microsoft JhengHei'] 
plt.rcParams['axes.unicode_minus'] = False
import numpy as np

# 正面/反面机率
probs = np.array([0.6, 0.4])
face = [0, 1]
plt.bar(face, probs)
plt.title('伯努利(Bernoulli)分配', fontsize=16)
plt.ylabel('机率', fontsize=14)
plt.xlabel('正面/反面', fontsize=14)
plt.ylim([0,1])

plt.show()
```

## 范例3. 绘制掷硬币10次的机率密度函数

```python
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['Microsoft JhengHei'] 
plt.rcParams['axes.unicode_minus'] = False
import numpy as np
from scipy.stats import binom

# 正面/反面机率
n = 10
p = 0.6

probs=[]
for k in range(11):
    p_binom = binom.pmf(k,n,p)
    probs.append(p_binom)

face=np.arange(11)
plt.bar(face, probs)
plt.title('二项(Binomial)分配', fontsize=16)
plt.ylabel('机率', fontsize=14)
plt.xlabel('正面次数', fontsize=14)

plt.show()
```

# 范例4. 绘制掷骰子10次的机率

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import multinomial

# 掷骰子次数
n = 6

# 假设掷出各种点数的机率
p = [0.1, 0.2, 0.2, 0.2, 0.2, 0.1]

probs=[]

# 掷骰子1~6点各出现1次的机率
p_multinom = multinomial(n,p)
p_multinom.pmf([1,1,1,1,1,1])
```

# 掷骰子60次，1~6点各出现10次的机率

```python
# 掷骰子60次
n = 60

# 掷骰子1~6点各出现10次的机率
p_multinom = multinomial(n,p)
p_multinom.pmf([10,10,10,10,10,10])
```

# 掷骰子1000次，1~6点各出现的次数

```python
# 随机抽样1000次
y = np.random.choice(np.arange(1,7), size=1000, p=p)

# np.bincount：0~6点各出现的次数，故要扣掉第一个元素
times = np.bincount(y)[1:]

# 绘图
plt.bar(np.arange(1,7), times)
plt.show()
```

# 范例5. 绘制各种λ值的卜瓦松分配机率密度函数

```python
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['Microsoft JhengHei'] 
plt.rcParams['axes.unicode_minus'] = False
import numpy as np
from scipy.stats import poisson

# λ = 2、4、6、8
for lambd in range(2, 10, 2):
    k = np.arange(0, 10)
    poisson1 = poisson.pmf(k, lambd)
    
    # 绘图
    plt.plot(k, poisson1, '-o', label=f"λ = {lambd}")

# 设定图形属性    
plt.xlabel('发生次数', fontsize=12)
plt.ylabel('机率', fontsize=12)
plt.title("Poisson Distribution (各种λ值)", fontsize=16)
plt.legend()
```
