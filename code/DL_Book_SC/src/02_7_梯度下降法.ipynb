# 梯度下降法(Gradient Descent)

## 范例1. 假定目标函数 f(x) = $x^2$，而非MSE，请使用梯度下降法求取最小值

```python
# 载入套件
import numpy as np
import matplotlib.pyplot as plt

# 目标函数(损失函数):y=x^2
def func(x): return x ** 2 #np.square(x)

# 目标函数的一阶导数:dy/dx=2*x
def dfunc(x): return 2 * x
```

```python
# 梯度下降
# x_start: x的起始点
# df: 目标函数的一阶导数
# epochs: 执行周期
# lr: 学习率
def GD(x_start, df, epochs, lr):    
    xs = np.zeros(epochs+1)    
    x = x_start    
    xs[0] = x    
    for i in range(epochs):         
        dx = df(x)        
        # x更新 x_new = x — learning_rate * gradient        
        x += - dx * lr         
        xs[i+1] = x    
    return xs
```

## 梯度下降法(Gradient Descent)示意图

```python
# 超参数(Hyperparameters)
x_start = 5     # 起始权重
epochs = 15     # 执行周期数 
lr = 0.3        # 学习率 

# 梯度下降法 
# *** Function 可以直接当参数传递 ***
w = GD(x_start, dfunc, epochs, lr=lr) 

color = 'r'    

plt.figure(figsize=(12,8))
t = np.arange(-6.0, 6.0, 0.01)
plt.plot(t, func(t), c='b')

# 设定中文字型
plt.rcParams['font.sans-serif'] = ['Microsoft JhengHei'] 
# 矫正负号
plt.rcParams['axes.unicode_minus'] = False

plt.title('梯度下降法', fontsize=20)
plt.xlabel('w', fontsize=20)
plt.ylabel('Loss', fontsize=20)

color = list('rgbymr')  # 切线颜色   
line_offset=2           #切线长度
for i in range(5, -1, -1):
    # 取相近两个点，画切线(tangent line)
    z=np.array([i+0.001, i])
    vec=np.vectorize(func)
    cls = np.polyfit(z, vec(z), deg=1)
    p = np.poly1d(cls)
    
    # 画切线
    x=np.array([i+line_offset, i-line_offset])
    y=np.array([(i+line_offset)*p[1]+p[0], (i-line_offset)*p[1]+p[0]])
    plt.plot(x, y, c=color[i-1])    

plt.show()
```

## 执行梯度下降法(Gradient Descent)

```python
# 超参数(Hyperparameters)
x_start = -5     # 起始权重
epochs = 15     # 执行周期数 
lr = 0.3        # 学习率 

# 梯度下降法 
# *** Function 可以直接当参数传递 ***
w = GD(x_start, dfunc, epochs, lr=lr) 
print (np.around(w, 2))

color = 'r'    
from numpy import arange
t = arange(-6.0, 6.0, 0.01)
plt.plot(t, func(t), c='b')
plt.plot(w, func(w), c=color, label='lr={}'.format(lr))    
plt.scatter(w, func(w), c=color, ) 

# 设定中文字型
plt.rcParams['font.sans-serif'] = ['Microsoft JhengHei'] 
# 矫正负号
plt.rcParams['axes.unicode_minus'] = False

plt.title('梯度下降法', fontsize=20)
plt.xlabel('X', fontsize=20)
plt.ylabel('损失函数', fontsize=20)
plt.show()
```

## 范例2. 假定损失函数如下，请使用梯度下降法求取最小值
$2x^4-3x-20$

```python
# 损失函数
def func(x): return 2*x**4-3*x+2*x-20

# 损失函数一阶导数
def dfunc(x): return 8*x**3-6*x+2
```

```python
from numpy import arange
t = arange(-6.0, 6.0, 0.01)
plt.plot(t, func(t), c='b')

# 设定中文字型
plt.rcParams['font.sans-serif'] = ['Microsoft JhengHei'] 
# 矫正负号
plt.rcParams['axes.unicode_minus'] = False

plt.title('梯度下降法', fontsize=20)
plt.xlabel('X', fontsize=20)
plt.ylabel('损失函数', fontsize=20)
plt.show()
```

## 执行梯度下降法(Gradient Descent)

```python
# 超参数(Hyperparameters)
x_start = 5     # 起始权重
epochs = 15000  # 执行周期数 
lr = 0.001      # 学习率 

# 梯度下降法 
# *** Function 可以直接当参数传递 ***
w = GD(x_start, dfunc, epochs, lr=lr) 
print (np.around(w, 2))

color = 'r'    
from numpy import arange
t = arange(-6.0, 6.0, 0.01)
plt.plot(t, func(t), c='b')
plt.plot(w, func(w), c=color, label='lr={}'.format(lr))    
plt.scatter(w, func(w), c=color, ) 

# 设定中文字型
plt.rcParams['font.sans-serif'] = ['Microsoft JhengHei'] 
# 矫正负号
plt.rcParams['axes.unicode_minus'] = False

plt.title('梯度下降法', fontsize=20)
plt.xlabel('X', fontsize=20)
plt.ylabel('损失函数', fontsize=20)
plt.show()
```

## lr = 0.3，会出现错误讯息，原因是学习率过大，梯度下降过程错过最小值，往函数左方逼近，造成损失函数值越来越大，最后导致溢位。

```python
# 超参数(Hyperparameters)
x_start = 5     # 起始权重
epochs = 15000  # 执行周期数 
lr = 0.3      # 学习率 

# 梯度下降法 
# *** Function 可以直接当参数传递 ***
w = GD(x_start, dfunc, epochs, lr=lr) 
print (np.around(w, 2))

color = 'r'    
from numpy import arange
t = arange(-6.0, 6.0, 0.01)
plt.plot(t, func(t), c='b')
plt.plot(w, func(w), c=color, label='lr={}'.format(lr))    
plt.scatter(w, func(w), c=color, ) 

# 设定中文字型
plt.rcParams['font.sans-serif'] = ['Microsoft JhengHei'] 
# 矫正负号
plt.rcParams['axes.unicode_minus'] = False

plt.title('梯度下降法', fontsize=20)
plt.xlabel('X', fontsize=20)
plt.ylabel('损失函数', fontsize=20)
plt.show()
```
