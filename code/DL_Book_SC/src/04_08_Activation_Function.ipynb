## 激励函数(Activation Function)

```python
import tensorflow as tf
from tensorflow.keras import activations
import numpy as np
import matplotlib.pyplot as plt
```

## ReLU(Rectified linear unit)

```python
# 设定 x = -10, -9, ..., 10 测试
x= np.linspace(-10, 10, 21)
x_tf = tf.constant(x, dtype = tf.float32)

# ReLU
y = activations.relu(x_tf).numpy()

# 绘图
plt.plot(x, y)
plt.show()
```

## ReLU 参数

```python
x= np.linspace(-10, 10, 21)
x_tf = tf.constant(x, dtype = tf.float32)
y = activations.relu(x_tf, threshold=5).numpy()

plt.plot(x, y)
plt.show()
```

```python
x= np.linspace(-10, 10, 21)
x_tf = tf.constant(x, dtype = tf.float32)
y = activations.relu(x_tf, max_value=5).numpy()

plt.plot(x, y)
plt.show()
```

```python
x= np.linspace(-10, 10, 21)
x_tf = tf.constant(x, dtype = tf.float32)
y = activations.relu(x_tf, alpha=0.5).numpy()
print(y)

plt.plot(x, y)
plt.show()
```

## Sigmoid

```python
x= np.linspace(-10, 10, 21)
x_tf = tf.constant(x, dtype = tf.float32)

# sigmoid
y = activations.sigmoid(x_tf).numpy()

# 模糊地带
plt.axvline(-4, color='r')
plt.axvline(4, color='r')

plt.plot(x, y)
plt.show()
```

## 使用NumPy计算Sigmoid

```python
x= np.linspace(-10, 10, 21)
y = 1 / (1 + np.e ** (-x))

plt.plot(x, y)
plt.show()
```

## tanh

```python
x= np.linspace(-10, 10, 21)
x_tf = tf.constant(x, dtype = tf.float32)

# tanh
y = activations.tanh(x_tf).numpy()

# 模糊地带
plt.axvline(-3, color='r')
plt.axvline(3, color='r')

plt.plot(x, y)
plt.show()
```

## Softmax

```python
# activations.softmax 输入需为 2 维资料，设定 x 为均匀分配，转换后每一列加总为 1
x = np.random.uniform(1, 10, 40).reshape(10, 4)
print('输入：\n', x)
x_tf = tf.constant(x, dtype = tf.float32)

# Softmax
y = activations.softmax(x_tf).numpy()
print('加总：', np.round(np.sum(y, axis=1)))
```

## 使用NumPy计算Softmax

```python
# 设定 x = 1, 2, ..., 10 测试
x= np.random.uniform(1, 10, 40)

# Softmax
y = np.e ** (x) / tf.reduce_sum(np.e ** (x))
print(sum(y))
```
