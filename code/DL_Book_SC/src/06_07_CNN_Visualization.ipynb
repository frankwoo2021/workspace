# 范例1. 重建卷积层处理后的影像
### 程式修改自
https://machinelearningmastery.com/how-to-visualize-filters-and-feature-maps-in-convolutional-neural-networks/

```python
# 载入套件
import tensorflow as tf
from tensorflow.keras.applications.vgg16 import VGG16
import matplotlib.pyplot as plt
import numpy as np
```

## 载入 VGG16 模型

```python
# 载入 VGG16 模型
model = VGG16()
model.summary()
```

## 取得所有卷积层

```python
# 取得所有卷积层
ixs = []
i=-1
for layer in model.layers:
    i+=1
    if 'conv' not in layer.name:
        continue
    print(i, layer.name)
    ixs.append(i)
```

## 视觉化特定层的特征图

```python
model.layers[1].name
```

```python
# 定义视觉化特征图的函数
def Visualize(layer_no=1, n_filters=6):
    # 取得权重(weight)
    filters, biases = model.layers[layer_no].get_weights()
    # 常态化(Normalization)
    f_min, f_max = filters.min(), filters.max()
    filters = (filters - f_min) / (f_max - f_min)
    
    # 绘制特征图
    ix = 1
    for i in range(n_filters):       
        f = filters[:, :, :, i]   # 取得每一个特征图        
        for j in range(3):        # 每列 3 张图 
            ax = plt.subplot(n_filters, 3, ix)  # 指定子视窗
            ax.set_xticks([])     # 无X轴刻度
            ax.set_yticks([])     # 无Y轴刻度
            plt.imshow(f[:, :, j], cmap='gray') # 以灰阶绘图
            ix += 1
    plt.show()
```

```python
Visualize(1)
```

```python
Visualize(15)
```

## 取得所有卷积层的特征图(Feature Map)

```python
i=-1
for layer in model.layers:
    i+=1
    # 取得所有卷积层
    if 'conv' not in layer.name:
        continue
    # 取得特征图
    filters, biases = layer.get_weights()
    print(i, layer.name, filters.shape)
```

## 重建第一个卷积层的输出图像

```python
# 设定第一个卷积层的输出为模型输出
model2 = tf.keras.models.Model(inputs=model.inputs, outputs=model.layers[1].output)

# 载入测试的图像
img = tf.keras.preprocessing.image.load_img('./images_test/bird.jpg', target_size=(224, 224))
img = tf.keras.preprocessing.image.img_to_array(img)    # 图像转为阵列
img = np.expand_dims(img, axis=0)                       # 加一维作为笔数
img = tf.keras.applications.vgg16.preprocess_input(img) # 前置处理(常态化)

# 预测
feature_maps = model2.predict(img)

# 将结果以 8x8 视窗显示
square = 8
ix = 1
plt.figure(figsize=(12,8))
for _ in range(square):
    for _ in range(square):
        ax = plt.subplot(square, square, ix)
        ax.set_xticks([])
        ax.set_yticks([])
        plt.imshow(feature_maps[0, :, :, ix-1], cmap='gray')
        ix += 1
plt.show()
```

## 重建 2, 5, 9, 13, 17 多层卷积层的输出图像

```python
# 取得 2, 5, 9, 13, 17 卷积层输出
ixs = [2, 5, 9, 13, 17]
outputs = [model.layers[i].output for i in ixs]
model2 = tf.keras.models.Model(inputs=model.inputs, outputs=outputs)

# 载入测试的图像
img = tf.keras.preprocessing.image.load_img('./images_test/bird.jpg', target_size=(224, 224))
img = tf.keras.preprocessing.image.img_to_array(img)    # 图像转为阵列
img = np.expand_dims(img, axis=0)                       # 加一维作为笔数
img = tf.keras.applications.vgg16.preprocess_input(img) # 前置处理(常态化)

# 预测
feature_maps = model2.predict(img)

# 将结果以 8x8 视窗显示
square = 8
for fmap in feature_maps:
    ix = 1
    plt.figure(figsize=(12,8))
    for _ in range(square):
        for _ in range(square):
            ax = plt.subplot(square, square, ix)
            ax.set_xticks([])
            ax.set_yticks([])
            plt.imshow(fmap[0, :, :, ix-1], cmap='gray')
            ix += 1
    plt.show()
```
