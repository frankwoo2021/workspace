## 常态分配的积分

```python
# 载入套件
import scipy.integrate as integrate
import numpy as np
import math


# 平均数(mean)、标准差(std)
mean = 0
std = 1
# 常态分配的机率密度函数(Probability Density Function, pdf)
f = lambda x: (1/((2*np.pi*std**2) ** .5)) * np.exp(-0.5*((x-mean)/std)**2) 

# 积分，从负无穷大至无穷大
i, e = integrate.quad(f, -np.inf, np.inf)

print('累积机率:', round(i, 2))
print('误差:', str(e))
```

## 计算常态分配的正负1、2、3倍标准差区间之机率

```python
# 1倍标准差区间之机率
i, e = integrate.quad(f, -1, 1)

print('累积机率:', round(i, 3))
print('误差:', str(e))
```

```python
# 2倍标准差区间之机率
i, e = integrate.quad(f, -2, 2)

print('累积机率:', round(i, 3))
print('误差:', str(e))
```

```python
# 3倍标准差区间之机率
i, e = integrate.quad(f, -3, 3)

print('累积机率:', round(i, 3))
print('误差:', str(e))
```

```python
# 95%信赖区间之机率
i, e = integrate.quad(f, -1.96, 1.96)

print('累积机率:', round(i, 3))
print('误差:', str(e))
```

## 使用随机乱数及直方图，绘制标准常态分配图

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# 使用 randn 产生标准常态分配的乱数 10000 个样本点
x = np.random.randn(10000)

# 直方图，参数 hist=False：不画阶梯直方图，只画平滑曲线
sns.distplot(x, hist=False)

# 设定图形属性
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()

# 对正负1、2、3倍标准差画虚线
plt.axvline(-3, c='r', linestyle=':')
plt.axvline(3, c='r', linestyle=':')
plt.axvline(-2, c='g', linestyle=':')
plt.axvline(2, c='g', linestyle=':')
plt.axvline(-1, c='b', linestyle=':')
plt.axvline(1, c='b', linestyle=':')

# 1 倍标准差机率
plt.annotate(text='', xy=(-1,0.25), xytext=(1,0.25), arrowprops=dict(arrowstyle='<->'))
plt.annotate(text='68.3%', xy=(0,0.26), xytext=(0,0.26))
# 2 倍标准差机率
plt.annotate(text='', xy=(-2,0.05), xytext=(2,0.05), arrowprops=dict(arrowstyle='<->'))
plt.annotate(text='95.4%', xy=(0,0.06), xytext=(0,0.06))
# 3 倍标准差机率
plt.annotate(text='', xy=(-3,0.01), xytext=(3,0.01), arrowprops=dict(arrowstyle='<->'))
plt.annotate(text='99.7%', xy=(0,0.02), xytext=(0,0.02))

plt.show()
```

## 范例1. 使用SciPy绘制机率密度函数(pdf)

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

# 观察值范围
z1, z2 = -4, 4

# 样本点
x = np.arange(z1, z2, 0.001) 
y = norm.pdf(x,0,1)

# 绘图
plt.plot(x,y,'black')

# 填色
plt.fill_between(x,y,0, alpha=0.3, color='b')
plt.show()
```

## 范例2. 绘制常态分配的正负1、2、3倍标准差区间及其机率

## 范例3. 使用SciPy绘制累积分配函数(cdf)

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

# 观察值范围
z1, z2 = -4, 4

# 样本点
x = np.arange(z1, z2, 0.001) 

# norm.cdf：累积分配函数
y = norm.cdf(x,0,1)

# 绘图
plt.plot(x,y,'black')

plt.show()
```

## 范例4. 使用SciPy绘制Student’s t 分配的机率密度函数(pdf)

```python
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['Microsoft JhengHei'] 
plt.rcParams['axes.unicode_minus'] = False
import numpy as np
from scipy.stats import t, norm

# 观察值范围
z1, z2 = -4, 4

# 样本点
x = np.arange(z1, z2, 0.1) 
dof = 10 #len(x) - 1
y = t.pdf(x, dof)

# 绘图
plt.plot(x,y,'red', label='t分配', linewidth=5)

# 绘制常态分配
y2 = norm.pdf(x)
plt.plot(x,y2, 'green', label='常态分配')

plt.legend()
plt.show()
```
